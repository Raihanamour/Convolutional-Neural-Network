```python
# import necessary modules

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications import VGG16
from tensorflow.keras.utils import to_categorical
```

```python
# Load cifar10

(X_train,y_train),(X_test,y_test) = cifar10.load_data()
```

```python
# pre-precossing

X_train = X_train/255.0
X_test = X_test/255.0

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Load pretrained model

basem = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
```

```python
# Freeze pre-trained layers

for layer in basem.layers:
  layer.trainable = False
```

```python
# Create new model on top

semodel = Sequential()
semodel.add(basem)
semodel.add(Flatten())
semodel.add(Dense(256,activation='relu'))
semodel.add(Dense(10,activation='softmax'))
```

```python
# Compile the model
semodel.compile(optimizer='adam',
                loss= 'categorical_crossentropy',
                metrics = ['accuracy'])
```

```python
# train the model

semodel.fit(X_train,y_train,epochs=15, batch_size=32,validation_data=(X_test,y_test))
```

```python
# Evaluate the model

y_test,y_pred =semodel.evaluate(X_test,y_test)
print('Test Accuracy: ',y_pred)
```
